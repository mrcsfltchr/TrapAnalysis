```python
import tifffile as tf
import numpy as np
from matplotlib import pyplot as plt
from fast_histogram import histogram1d
```

```python
tif = tf.TiffFile('./181008_CecB-PCPG vesicles- after flushing_1_MMStack_Pos22.ome.tif')
```

```python
drug_arrived_frame = tif.asarray(key = 100)
```

```python
#flatten the image into a 1d array ( only interested in one variable, the intensity)
intensities = drug_arrived_frame.flatten()
xmin = np.min(intensities)
xmax = np.max(intensities)

hist = histogram1d(x=intensities,bins = 40,range=(xmin,xmax))
```

```python
hist
```

```python
plt.bar(x = range(1,41),height=hist)
```
