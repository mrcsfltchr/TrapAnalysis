```python
comparison = data[data.columns[[27,-6]]].to_numpy()
```

```python
normalised_data = np.zeros_like(comparison)
for col in range(0,comparison.shape[1]):
    normalised_data[:,col] = comparison[:,col]/np.max(comparison[:,col][:70])
```

```python
%matplotlib qt
matplotlib.rcParams.update({'xtick.major.size':6}) #length of ticks
matplotlib.rcParams.update({'ytick.major.size':6})
matplotlib.rcParams.update({'lines.markeredgewidth':2}) #width of ticks
matplotlib.rcParams.update({'lines.markersize':6}) 


fig,ax = plt.subplots(1,1)
plt.style.use('classic')
fig.set_facecolor('w')

ax.plot(0.5*np.arange(70),normalised_data[:,0][:70],c = 'r')
ax.plot(0.5*np.arange(70),normalised_data[:,1][150:220],c = 'k')
```
